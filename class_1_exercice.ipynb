```python
plt.figure(figsize=(10, 10))

target_x_val = 2
target_y_val = 4
target_r_val = 1

xy = src.point_data.generate_circle_points(
    x_mu=target_x_val,
    x_std=0.1,
    y_mu=target_y_val,
    y_std=0.1,
    r_mu=target_r_val,
    r_std=0.1,
    num_pts=100
)

plt.scatter(xy[:, 0], xy[:, 1])
plt.grid()

plt.axhline(0, color="black")
plt.axvline(0, color="black")

plt.gca().add_patch(plt.Circle(
    (target_x_val, target_y_val),
    target_r_val,
    color="gray",
    linestyle="--",
    fill=False
))

plt.scatter(
    [target_x_val],
    [target_y_val],
    color="gray"
)

#

#max_plt_x = abs(target_x_val) + target_r_val * 1.5
#max_plt_y = abs(target_y_val) + target_r_val * 1.5
#max_plt_s = max(max_plt_y, max_plt_y)

#plt.xlim(-max_plt_s, max_plt_s)
#plt.ylim(-max_plt_s, max_plt_s)

plt.xlim(target_x_val - (target_r_val * 1.5), target_x_val + (target_r_val * 1.5))
plt.ylim(target_y_val - (target_r_val * 1.5), target_y_val + (target_r_val * 1.5))

#

plt.show()
```
